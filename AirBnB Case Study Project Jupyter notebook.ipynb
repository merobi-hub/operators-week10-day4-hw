```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?
hoods = air_bnb['neighbourhood_group']
hoods.head()
total = max(range(len(hoods)))
print(total)
# Fun to find the neighborhood with most listings (and # of listings)
def mostHoods(df):
    hood_stats = {}
    for hood in hoods:
        if hood in hood_stats.keys():
            hood_stats[hood] += 1
        else:
            hood_stats[hood] = 1
    for k,v in hood_stats.items():
        if v == max(hood_stats.values()):
            return k, v
print(mostHoods(hoods))
# Fun to count the num of diff neighborhoods
def numHoods(df):
    hood_stats = {}
    for hood in hoods:
        if hood in hood_stats.keys():
            hood_stats[hood] += 1
        else:
            hood_stats[hood] = 1
    nh = ''
    count = 0
    for hood in hood_stats.keys():
        if hood == nh:
            continue
        else:
            count += 1
            nh = hood
    return count
print(numHoods(hoods))
```

```python
# Are private rooms the most popular in manhattan?
rooms = air_bnb[['neighbourhood_group', 'room_type']]

private = rooms[rooms['room_type'].str.contains("Private")]
privateHoods = private['neighbourhood_group']

def mostPrivateRooms(df):
    hood_stats = {}
    for hood in df:
        if hood in hood_stats.keys():
            hood_stats[hood] += 1
        else:
            hood_stats[hood] = 1
    for k,v in hood_stats.items():
        if v == max(hood_stats.values()):
            return f'Private rooms are most popular in {k}: {v}'
print(mostPrivateRooms(privateHoods))
```
